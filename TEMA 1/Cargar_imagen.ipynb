```python
import cv2
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Cargar la imagen en color
imagen = cv2.imread('flores_amarillas.jpg')  # Se carga en formato BGR por defecto

# 2Ô∏è‚É£ Separar los canales de la imagen
B, G, R = cv2.split(imagen)  # OpenCV usa BGR, no RGB

# 3Ô∏è‚É£ Visualizar los canales por separado con Matplotlib
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

# Imagen original (Convertimos de BGR a RGB para que Matplotlib la muestre correctamente)
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
axs[0].imshow(imagen_rgb)
axs[0].set_title('Imagen Original (RGB)')
axs[0].axis('off')

# Canal Azul (B)
axs[1].imshow(B, cmap='Blues')  
axs[1].set_title('Canal Azul (B)')
axs[1].axis('off')

# Canal Verde (G)
axs[2].imshow(G, cmap='Greens')  
axs[2].set_title('Canal Verde (G)')
axs[2].axis('off')

# Canal Rojo (R)
axs[3].imshow(R, cmap='Reds')  
axs[3].set_title('Canal Rojo (R)')
axs[3].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Cargar la imagen en color
imagen = cv2.imread('flores_amarillas.jpg')  # OpenCV carga en formato BGR

# 2Ô∏è‚É£ Separar los canales B, G, R
B, G, R = cv2.split(imagen)

# 3Ô∏è‚É£ Aplicar la f√≥rmula manualmente para convertir a escala de grises
imagen_gris_manual = (0.299 * R + 0.587 * G + 0.114 * B).astype(np.uint8)

# 4Ô∏è‚É£ Convertir la imagen a escala de grises directamente con OpenCV
imagen_gris_opencv = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# 5Ô∏è‚É£ Comparar las dos im√°genes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Imagen original en RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
axs[0].imshow(imagen_rgb)
axs[0].set_title('Imagen Original (RGB)')
axs[0].axis('off')

# Imagen en escala de grises (M√©todo Manual)
axs[1].imshow(imagen_gris_manual, cmap='gray')
axs[1].set_title('Escala de Grises - M√©todo Manual')
axs[1].axis('off')

# Imagen en escala de grises (OpenCV)
axs[2].imshow(imagen_gris_opencv, cmap='gray')
axs[2].set_title('Escala de Grises - OpenCV')
axs[2].axis('off')

plt.show()

# 6Ô∏è‚É£ Comparar si las im√°genes son iguales
iguales = np.array_equal(imagen_gris_manual, imagen_gris_opencv)
print(f"¬øLas im√°genes son iguales? {iguales}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Cargar la imagen en escala de grises
imagen = cv2.imread('flores_amarillas.jpg', cv2.IMREAD_GRAYSCALE)

# 2Ô∏è‚É£ Calcular el histograma
histograma = cv2.calcHist([imagen], [0], None, [256], [0, 256])

# 3Ô∏è‚É£ Visualizar la imagen y su histograma
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Mostrar la imagen en escala de grises
axs[0].imshow(imagen, cmap='gray')
axs[0].set_title('Imagen en Escala de Grises')
axs[0].axis('off')

# Mostrar el histograma
axs[1].plot(histograma, color='black')
axs[1].set_title('Histograma de Intensidad')
axs[1].set_xlabel('Valor de p√≠xel')
axs[1].set_ylabel('Frecuencia')
axs[1].grid()

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Cargar la imagen en escala de grises
imagen = cv2.imread('alto_contraste.jpeg', cv2.IMREAD_GRAYSCALE)

# 2Ô∏è‚É£ Definir los par√°metros de transformaci√≥n
alpha = 155/255  # Factor de contraste (>1 aumenta contraste, <1 lo reduce)
beta = 100    # Factor de brillo (>0 aclara, <0 oscurece)

# 3Ô∏è‚É£ Aplicar la transformaci√≥n lineal: T(r) = Œ± * r + Œ≤
imagen_transformada = cv2.convertScaleAbs(imagen, alpha=alpha, beta=beta)

# 4Ô∏è‚É£ Visualizar la imagen original y la transformada
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Imagen original
axs[0].imshow(imagen, cmap='gray')
axs[0].set_title('Imagen Original')
axs[0].axis('off')

# Imagen transformada
axs[1].imshow(imagen_transformada, cmap='gray')
axs[1].set_title(f'Imagen Transformada (Œ±={alpha}, Œ≤={beta})')
axs[1].axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Cargar la imagen en escala de grises
imagen = cv2.imread('imagen.jpg', cv2.IMREAD_GRAYSCALE)

# 2Ô∏è‚É£ Calcular la constante 'c' para normalizar la imagen a 255
c = 255 / np.log(1 + np.max(imagen))

# 3Ô∏è‚É£ Aplicar la transformaci√≥n logar√≠tmica: T(r) = c * log(1 + r)
imagen_log = c * np.log(1 + imagen.astype(np.float32))

# 4Ô∏è‚É£ Normalizar la imagen y convertir a uint8
imagen_log = np.clip(imagen_log, 0, 255).astype(np.uint8)

# 5Ô∏è‚É£ Visualizar las im√°genes
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Imagen original
axs[0].imshow(imagen, cmap='gray')
axs[0].set_title('Imagen Original')
axs[0].axis('off')

# Imagen transformada
axs[1].imshow(imagen_log, cmap='gray')
axs[1].set_title('Transformaci√≥n Logar√≠tmica')
axs[1].axis('off')

plt.show()
```
