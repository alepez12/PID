## Ejercicio: Uso de la función `dilate()` de OpenCV

### Descripción:
Usaremos la función **`cv2.dilate()`** de OpenCV para hacer dilataciones de una imagen. Se aplicará un **elemento estructural** definido por el usuario y se explorará cómo este afecta a la imagen.

### Tareas:

**a)** Usa la función de OpenCV **`dilate()`** para hacer dilataciones de una imagen. 

**b)** Define antes el **elemento estructural**. Averigua las opciones que da OpenCV para definirlo.

**c)** Averigua qué ocurre si aplicas la dilatación a una imagen en escala de grises.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
image = cv2.imread('j.png', cv2.IMREAD_GRAYSCALE)

# Crear un elemento estructural
kernel = np.ones((5, 5), np.uint8)  # Elemento estructural 5x5

# Aplicar dilatación
dilated_image = cv2.dilate(image, kernel, iterations=1)

# Mostrar la imagen original y la dilatada
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Imagen Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_image, cmap='gray')
plt.title("Imagen Dilatada")
plt.axis('off')

plt.show()
```

## Ejercicio: Uso de la función `morphologyEx()` de OpenCV con `cv.MORPH_OPEN`

### Descripción:
Usaremos la función **`cv2.morphologyEx()`** de OpenCV con el argumento **`cv2.MORPH_OPEN`** para hacer la **apertura** de una imagen. La operación de apertura es una combinación de erosión seguida de dilatación. Esta operación elimina el ruido y reduce las pequeñas áreas brillantes en la imagen.

### Tareas:

**a)** Usa la función de OpenCV **`morphologyEx()`** con el argumento **`cv2.MORPH_OPEN`** para hacer apertura de una imagen.

**b)** Define antes el **elemento estructural**. Averigua las opciones que da OpenCV para definirlo.

**c)** Averigua qué ocurre si aplicas la apertura a una imagen en escala de grises.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Crear una imagen de 100x100 píxeles con un fondo negro
image = np.zeros((100, 100), dtype=np.uint8)

# Añadir puntos blancos (ruido) a la imagen
cv2.circle(image, (20, 20), 5, 255, -1)  # Añadir un círculo blanco
cv2.circle(image, (80, 80), 5, 255, -1)  # Añadir otro círculo blanco
cv2.circle(image, (50, 50), 5, 255, -1)  # Otro círculo blanco

# Crear un elemento estructural (kernel) 5x5
kernel = np.ones((5, 5), np.uint8)

# Aplicar apertura a la imagen
opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Mostrar la imagen original y la imagen con apertura
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Imagen Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(opened_image, cmap='gray')
plt.title("Imagen con Apertura")
plt.axis('off')

plt.show()
```

## Ejercicio 1: Eliminación de ruido

### Descripción:
En este ejercicio se pretende eliminar el ruido de la imagen **circulos_ruido.png** usando **operaciones morfológicas binarias**. Se deben aplicar las mejores operaciones morfológicas con los elementos estructurales más pequeños posibles, para obtener el menor error posible. Se debe calcular el **Error Cuadrático Medio (MSE)** comparando la imagen procesada tras eliminar el ruido con la imagen original para corroborar la mejor elección.

Además, se debe aplicar una operación morfológica para eliminar los pequeños círculos negros dentro de los círculos blancos en la imagen sin ruido, minimizando el error.

### Tareas:

**a)** Dada la imagen **circulos_ruido.png**, se pretende eliminar el ruido usando operaciones morfológicas binarias. Razona cuál será la mejor secuencia de operaciones para ello con los mejores elementos estructurales (que podrían ser distintos para eliminar un ruido y otro). Queremos que los elementos estructurales sean lo más pequeños posibles para obtener el menor error posible. Calcula el Error Cuadrático Medio comparando la imagen procesada tras eliminar el ruido con la imagen original para corroborar la mejor elección.

**b)** Sobre la imagen sin ruido obtenida, determinar la mejor operación morfológica para eliminar los pequeños círculos negros dentro de los blancos cometiendo un error lo más pequeño posible.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen con ruido
image = cv2.imread('circulos_ruidos.png', cv2.IMREAD_GRAYSCALE)

# **a) Eliminar el ruido usando operaciones morfológicas binarias**

# Crear el elemento estructural para eliminar el ruido (usaremos un elemento pequeño)
kernel = np.ones((3, 3), np.uint8)  # Elemento estructural 3x3

# Operación de apertura para eliminar el ruido pequeño (erosión seguida de dilatación)
image_cleaned = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Mostrar imagen original y procesada
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Imagen Original con Ruido")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_cleaned, cmap='gray')
plt.title("Imagen Sin Ruido")
plt.axis('off')

plt.show()

# **Cálculo del Error Cuadrático Medio (MSE)**
def mse(image1, image2):
    return np.mean((image1 - image2) ** 2)

# Comparar la imagen original con la imagen procesada (sin ruido)
error_mse = mse(image, image_cleaned)
print(f"Error Cuadrático Medio (MSE) entre la imagen original y la imagen sin ruido: {error_mse}")

# **b) Eliminar los pequeños círculos negros dentro de los blancos en la imagen sin ruido**

# Usar una operación morfológica para eliminar los círculos negros dentro de los blancos
kernel2 = np.ones((5, 5), np.uint8)  # Elemento estructural más grande

# Operación de apertura para eliminar los pequeños círculos negros (erosión seguida de dilatación)
image_no_black_circles = cv2.morphologyEx(image_cleaned, cv2.MORPH_OPEN, kernel2)

# Mostrar la imagen sin los círculos negros
plt.imshow(image_no_black_circles, cmap='gray')
plt.title("Imagen Sin Círculos Negros")
plt.axis('off')
plt.show()

# Calcular el MSE entre la imagen procesada sin círculos negros y la original sin ruido
error_mse_no_black_circles = mse(image_cleaned, image_no_black_circles)
print(f"Error Cuadrático Medio (MSE) entre la imagen sin ruido y la imagen sin círculos negros: {error_mse_no_black_circles}")
```

## Ejercicio 2: Extracción de fronteras

### Descripción:
Dada una imagen binaria **I**, los píxeles de la frontera de **I** se obtienen restando a **I** el resultado de dilatar la imagen con un **elemento estructural blanco 3x3**. Se debe definir una función llamada **`extraer_fronteras()`** que realice este proceso y luego aplicar esta función a la imagen **pez.png**.

### Tareas:

**a)** Define una función llamada **`extraer_fronteras()`** que realice el proceso descrito de extracción de fronteras de la imagen binaria.

**b)** Aplica la función **`extraer_fronteras()`** a la imagen **pez.png**.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Función para extraer las fronteras de una imagen binaria
def extraer_fronteras(image):
    # Crear un elemento estructural 3x3 blanco
    kernel = np.ones((3, 3), np.uint8)
    
    # Dilatar la imagen
    dilated_image = cv2.dilate(image, kernel, iterations=1)
    
    # Extraer las fronteras restando la imagen dilatada de la imagen original
    fronteras = dilated_image - image
    
    return fronteras

# Cargar la imagen 'pez.png' en escala de grises
image = cv2.imread('pez.png', cv2.IMREAD_GRAYSCALE)

# Convertir la imagen a binaria (suponiendo que la imagen es en escala de grises)
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Extraer las fronteras de la imagen binaria
fronteras = extraer_fronteras(binary_image)

# Mostrar la imagen binaria original y la imagen de fronteras
plt.subplot(1, 2, 1)
plt.imshow(binary_image, cmap='gray')
plt.title("Imagen Binaria")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(fronteras, cmap='gray')
plt.title("Fronteras Extraídas")
plt.axis('off')

plt.show()
```
